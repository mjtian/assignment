# 简单线条及坐标轴设置

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 50)
y1 = 2*x + 1
y2 = x**2
plt.figure()
plt.plot(x, y2)
plt.show()

plt.figure(num=3, figsize=(8, 5),)
plt.plot(x, y2)
plt.plot(x, y1,color='red',linewidth=1.0,linestyle='--')
plt.xlim((-1,2)) #设置x轴坐标范围
plt.ylim((-2,3)) #设置y轴的坐标范围
plt.xlabel('i am x') #设置x轴的名称
plt.ylabel('i am y') #设置y轴的名称
new_ticks=np.linspace(-1,2,5) #重新确定坐标的范围以及个数
print(new_ticks) #打印出新坐标
plt.xticks(new_ticks) #讲新的坐标设置成x的坐标
plt.yticks([-2,-1.8,-1,1.22,3],[r'$really\ bad$',r'$bad$',r'$normal$',r'$good$',r'$really\ good$'])
#y轴刻度设置名称 正则表达式可以直接写公式
plt.show()
```
